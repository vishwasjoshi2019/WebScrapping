```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# ... (other imports and initialization)

# Lists to store company names, addresses, and URLs
company_names = []
company_address = []
company_urls = []

# Iterate through the pages (1 to 7)
for i in range(1, 8):
    html = scraper.get("https://www.designrush.com/agency/bpo/us" + "?page=" + str(i)).content
    soup = BeautifulSoup(html, 'html.parser')

    # Extract company names and addresses
    company_name_elements = soup.select("html > body > div:nth-of-type(2) > div:nth-of-type(2) > main > div > div:nth-of-type(2) > div > ul > li > div:nth-of-type(1) > div:nth-of-type(1) > div > div:nth-of-type(1) > h3 > a")
    company_address_elements = soup.select("html > body > div:nth-of-type(2) > div:nth-of-type(2) > main > div > div:nth-of-type(2) > div > ul > li > div:nth-of-type(1) > div:nth-of-type(3) > span:nth-of-type(1)")

    # Extract company URLs using the specified XPath
    company_url_elements = soup.select("html > body > div:nth-of-type(2) > div:nth-of-type(2) > main > div > div:nth-of-type(2) > div > ul > li > div:nth-of-type(2) > a")

    # Extract the text from the elements
    company_names += [element.text for element in company_name_elements]
    company_address += [element.text for element in company_address_elements]
    company_urls += [element['href'] for element in company_url_elements]

# Create a DataFrame
data = {
    'Company Name': company_names,
    'Company Address': company_address,
    'Company URL': company_urls
}

df = pd.DataFrame(data)
print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

def extract_contact_info(url):
    contact_info = {
        'Email': None,
        'Phone': None
    }

    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')

        # Extract email addresses from href attributes that contain "mailto"
        email_from_href_pattern = re.compile(r'(?<=href="mailto:)([A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,})(?=")')
        email_from_href_match = email_from_href_pattern.search(response.text)
        if email_from_href_match:
            contact_info['Email'] = email_from_href_match.group()
        else:
            # If no email found in href attributes, use the initial email regex
            email_pattern = re.compile(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b')
            email_match = email_pattern.search(response.text)
            if email_match:
                contact_info['Email'] = email_match.group()

        # Extract phone numbers from href attributes that contain "tele"
        phone_pattern = re.compile(r'(?<=href="tel:)(\+?[0-9()\s-]+)(?=")')
        phone_matches = re.findall(phone_pattern, response.text)
        if phone_matches:
            contact_info['Phone'] = phone_matches
        else:
            # If no phone number found in href attributes, use another regex for US phone number matching
            us_phone_pattern = re.compile(r'\b(?:\+?1[-.])?\(?\d{3}\)?[-.\s]?\d{3}[-.\s]?\d{4}\b')
            us_phone_match = re.search(us_phone_pattern, response.text)
            if us_phone_match:
                contact_info['Phone'] = us_phone_match.group()

    except requests.exceptions.RequestException as e:
        print(f"Error accessing {url}: {e}")

    return contact_info


# Iterate through each URL in the DataFrame and extract contact information
df['Contact Info'] = df['Company URL'].apply(extract_contact_info)

# Split the 'Contact Info' dictionary into separate columns
# df[['Email', 'Phone']] = pd.DataFrame(df['Contact Info'].tolist(), index=df.index)

# # Drop the 'Contact Info' column
# df.drop('Contact Info', axis=1, inplace=True)

# Display the updated DataFrame
print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import time

# URL of the website
base_url = "https://clutch.co/us/bpo?page="
company_names = []
company_addresses = []

# Mimic a user-agent to avoid being blocked
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 '
                  '(KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

# Loop through pages from 1 to 44
for page_number in range(1, 45):
    url = base_url + str(page_number)
    response = requests.get(url, headers=headers)
    if response.status_code != 200:
        print(f"Failed to retrieve page {page_number}")
        continue

    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all company name elements using XPath
    name_elements = soup.select("html > body > main > section > div:nth-of-type(2) > ul > li > div > div:nth-of-type(1) > div:nth-of-type(1) > div > h3 > a")

    for name_element in name_elements:
        company_names.append(name_element.text)

    # Find all company address elements using XPath
    address_elements = soup.select("html > body > main > section > div:nth-of-type(2) > ul > li > div > div:nth-of-type(1) > div:nth-of-type(2) > div:nth-of-type(1) > div:nth-of-type(2) > div:nth-of-type(4) > span")

    for address_element in address_elements:
        company_addresses.append(address_element.text)

    # Add a delay between requests to mimic human behavior
    time.sleep(5)  # Adjust the delay time as needed

# Print the collected company names and addresses
for i in range(len(company_names)):
    print(f"Company Name: {company_names[i]}")
    print(f"Company Address: {company_addresses[i]}")
    print("=" * 50)
```
